### 2. 버퍼에 데이터 붙이기

```python
gid1 = lpg_buffer.merge(df,on='gid',how='inner')
```

```python
gid1
```

```python
address=gid1['소재지'].unique()
address
```

```python
for index,i in enumerate(address):
    globals()['LPG_{}'.format(index)] = gid1[gid1['소재지']==i].iloc[:,:-1]
```
