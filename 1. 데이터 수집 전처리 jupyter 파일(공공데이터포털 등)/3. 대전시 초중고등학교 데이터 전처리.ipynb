### 1. csv파일 geojson파일로 변환하기

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point, Polygon, LineString
```

```python
df = pd.read_csv('C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/2.사용_데이터/3. 공공데이터포털/한국교원대학교_초중등학교위치_20210325_1616658916059_2895709.csv',encoding='cp949')
```

### 2. 대전광역시 데이터만 추출

```python
df['시도교육청명'].value_counts()
```

```python
daejeon=df[df['시도교육청명']=='대전광역시교육청']
```

### 3. dataframe -> geojson 변환

```python
daejeon
```

```python
daejeon['geometry'] = daejeon.apply(lambda row : Point([row['경도'], row['위도']]), axis=1)
daejeon = gpd.GeoDataFrame(daejeon, geometry ='geometry')
```

```python
daejeon['geometry']
```

```python
daejeon_geojson = gpd.GeoDataFrame(daejeon)
daejeon_geojson.to_file('대전_초중고.geojson', driver="GeoJSON")
```
