### 2. 데이터 동 별로 count 하기

```python
fire_count_dong_17=pd.DataFrame(fire_17['읍면동'].value_counts())
```

```python
fire_count_dong_17.reset_index(inplace = True)
fire_count_dong_17.rename(columns = {"index":"EMD_KOR_NM","읍면동":"count"}, inplace = True)
```

```python
fire_count_dong_17
```

### 3. 대전광역시_법정경계(읍면동)별 17년도 화재 발생 count geojson 만들기

```python
daejeon_dong_fire_count_17=daejeon_dong.merge(fire_count_dong_17,how="outer",on="EMD_KOR_NM")
```

```python
daejeon_dong_fire_count_17
```

```python
daejeon_dong_fire_count_17 = gpd.GeoDataFrame(daejeon_dong_fire_count_17)
daejeon_dong_fire_count_17.to_file('C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/2.사용_데이터(정제데이터)/1.QGIS에 사용할 데이터(gid 가능데이터)/17년도 동별 화재 count.geojson', driver='GeoJSON')
```
