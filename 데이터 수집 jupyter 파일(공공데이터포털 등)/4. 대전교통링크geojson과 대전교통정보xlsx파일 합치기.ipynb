```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

### 1. 데이터 불러오기

```python
traffic_link=gpd.read_file("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/2.사용_데이터(정제데이터)/2.[대전]COMPAS공개데이터/17.대전광역시_교통링크(2018).geojson")
daejeon_traffic = pd.read_excel("C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/2.사용_데이터(정제데이터)/3. 공공데이터포털(api, csv 둘다)/대전교통정보.xlsx")
```

```python
## daejeon_traffic의 linkID를 LINK_ID로 바꾸기
daejeon_traffic.rename(columns={"linkID":"LINK_ID"},inplace=True)
```

```python
## traffic_link 의 LINK_ID 열속성을 object -> int64로 바꾸기
traffic_link['LINK_ID'] = pd.to_numeric(traffic_link['LINK_ID'])
```

### 2. 데이터 합치기

```python
traffic = traffic_link.merge(daejeon_traffic,how='inner',on="LINK_ID")
```

```python
traffic
```

### 3. geojson으로 저장

```python
traffic_geojson = gpd.GeoDataFrame(traffic)
traffic_geojson.to_file('C:/Users/Kim Won Gyeung/Desktop/공모전/2021 빅데이터 분석 및 아이디어 공모전/2.사용_데이터(정제데이터)/1.QGIS에 사용할 데이터(gid 가능데이터)/대전_교통링크별_교통량.geojson', driver='GeoJSON')
```
